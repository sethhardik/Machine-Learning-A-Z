# SVM 
    SVM STANDS FOR SUPPORT VECTOR MACHINE
    IT CAN BE USED FOR BOTH CLASSIFICATION AND REGRESSION BUT MAINLY WE USE IT FOR CLASSIFICATION
    
## WHAT IS SVM ?

     IT ANALYSIS THE DATA AND RECOGINIZE THE PATTERN 
     IT IS A REPRESENTAION OF THE EXAMPLES AS POINT IN SPACE, MAPPED SO THAT THE EXAMPLE OF THE SEPERATED CATEGORIES ARE DIVIDED BY A CLEAR GAP
     
     IT CREATES A *HYPER PLANE* THAT SEPERATES CLASS 1 AND CLASS 2

### WHAT IS HYPERPLANE ?
    
     IT IS A LINEAR DECISION SURFACE THAT SPLITS THE SPACE INTO 2 PARTS 
     IT IS BINARY CLASSIFIER
     DIMENSION =N-2        THAT IS FOR 2D PLANE HYPERPLANE IS 1D

#### WHAT IS SUPPORT VECTOR ?
     
      SVM CHOOSES THE OPTIMAL HYPERPLANE AND MAXIMIZE THE MARGIN B/W THE 2 CLASS
      SUPPORT VECTOR ARE THE VECTOR POINTS AT THE MARGIN LINE 
      
      
  SVM CAN BE USED FOR BOTH LINEAR AND NON LINEAR CLASSIFICTAION.
       
       FOR NON LINEAR WE USE KERNAL TRICK TO MAKE A NON LINEAR HYPERPLANE
       
       
##### LOSS FUNCTION USED IS HINGE LOSS

##### OBJETCIVE FUNCTION USED IS COMBINATION OF REGULARIZER + TOTAL LOSS FUNCTION 

##### WHAT IS REGULARIZER FUNCTION ?
   
     REGULIZER=min(LAMBDA)||W||^2
     REGULIZER TOO HIGH -----> OVERFIT
     REGULIZER TOO LOW  ----->UNDERFIT
    
###### OBJECTIVE FUNCTION OPTIMZATION 

     DO GRADIENT  DESENT BY PERFORMING PARTIAL DERIVATIVE OF OBJECTIVE FUNCTION 
     
     WEIGHT=WEIGHT+LEARNING RATE(-2(LAMBDA)(WEIGHT))
    
    
###### MAIN HYPERPARAMETER IN SVM ARE KERNAL, C, GAMMA


# uses for small dataset were linear regression can prove to be expensive

```python
from sklearn import datasets
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
cancer=datasets.load_breast_cancer()
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df.head()
```

```python
df["type"]=pd.Categorical.from_codes(cancer.target,cancer.target_names)
df.head(10)
```

```python
data=cancer.data[:,:]
target=cancer.target
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data,target,test_size=0.2,random_state=42)
xtrain.shape
```

```python
svm=SVC()
```

```python
svm.fit(xtrain,ytrain)
```

```python
predict=svm.predict(xtest)
print("Accuracy score:",accuracy_score(predict,ytest))
```

```python
# now lets see the if accuracy can be improved by hyperparameter optimization
```

```python
#grid seach takes a dictitionary
parameter={"C":[1,0.1,10,100],"gamma":[1,0.1,0.01,0.001,]}
```

```python
grid=GridSearchCV(SVC(),parameter,verbose=1)
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_         # best parameter is given now we can use these values in the model
```

```python
svm2=SVC(C=10,gamma=0.001)
```

```python
svm2.fit(xtrain,ytrain)
```

```python
predict_y=svm2.predict(xtest)
```

```python
print("accuracy score:",accuracy_score(predict_y,ytest))
print("confussion matrix:",confusion_matrix(predict_y,ytest))
```

```python
# clearly the model improved way too much 
```
